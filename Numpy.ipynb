```python
#rint():to get the round values of array or a matrix
b=np.array([1.5,2.6,3.3,4.5,5.9,6.1,7.6])
print(np.rint(b))
#true_divide():dividing the first array with second array and gives  the exact values as output
b=np.array([10,25,36,41,52])
a=np.array([1,2,3,4,5])
print(np.true_divide(b,a))
#unique():removes the duplicate values
b=np.array([1,1,2,2,3,3,4,4])
print(np.unique(b))
#union1d():gives 1D list with deletion of duplicate values
a=np.array([[1,2],[3,4]])
b=np.array([[1,2],[7,8]])
print(np.union1d(a,b))
#intersect1d():prints only common elements
a=np.array([[1,2],[3,4]])
b=np.array([[1,2],[7,8]])
print(np.intersect1d(a,b))
#setdiff1d:
a=np.array([[1,2],[3,4]])
b=np.array([[1,2],[7,8]])
print(np.setdiff1d(a,b))
#hypot():
a=8
b=6
c=np.hypot(a,b)
print(c)
#sin()
x=np.sin(0)
print(x)
```

```python
#divmod:prints both quotient and remainder
a=np.array([[10,21,30,40,50,60]])
b=np.array([[20,21,2,20,25,25]])
print(np.divmod(a,b))
#mod:gives remainder
a=np.array([[10,21,30,40,50,60]])
b=np.array([[20,21,2,20,25,25]])
print(np.mod(a,b))
#divide:gives exact vale of quotient
a=np.array([[10,21,30,40,50,60]])
b=np.array([[20,21,2,20,25,25]])
print(np.divide(a,b))
#multiply:element wise multiplication
a=np.array([[1,2],[3,4]])
b=np.array([[1,2],[3,4]])
print(np.multiply(a,b))
```

```python
from numpy import random
#normal distrubtion
x=random.normal(size=(2,3))
print(x)
#loc-mean;sacle-SD;
x=random.normal(loc=1,scale=2,size=(2,3))
print(x)
#Binomal Distribution:
x=random.binomial(n=12,p=0.5,size=10) # n is n.of trails,p is probability
print(x)
#poisson distribution
x=random.poisson(lam=2,size=10)
print(x)
x=random.choice([1,2,3,4]) #prints any element randomly
print(x)
x=random.choice([3,5],p=[0.5,0.5],size=(3,5))
print(x)
```

```python
import matplotlib.pyplot as plt
a = [230,560,780,127,128]
b = [200,160,270,127,400]
years = [1,2,3,4]
profit_a = [(a[i]-a[i-1]) for i in range(1,len(a))]
profit_b = [(b[i]-b[i-1]) for i in range(1,len(b))]
plt.subplot(2,1,2) #gives two subplots for givn two
plt.plot(years,profit_a,color='hotpink',linewidth = '3',label =
'CompanyA',marker='>',ms='15',mec='k')
plt.subplot(2,1,1)
plt.plot(years,profit_b,color='black',linestyle='dotted',label =
'CompanyB',marker = 'H')
plt.show()
```
